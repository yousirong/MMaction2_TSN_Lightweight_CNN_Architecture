```python
import pandas as pd
from glob import glob

file_names = glob("./assault_label_*.csv")
total = pd.DataFrame()

for file_name in file_names:
    temp = pd.read_csv(file_name, sep=',', encoding='utf-8')
    total = pd.concat([total, temp])

total.to_csv("./assault_label_total.csv",index=False)
```

```python
file_names = glob("./train_label_*.csv")
total = pd.DataFrame()

for file_name in file_names:
    temp = pd.read_csv(file_name, sep=',', encoding='utf-8')
    total = pd.concat([total, temp])

total.to_csv("./train_label_total.csv",index=False)
```
